```python
import numpy as np
import matplotlib.pyplot as plt

def model(x):
    return 20.6 + 0.4945 * x

x = np.linspace(0, 60, 1000)
y = model(x)

np.random.seed(42)
x_data = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])
y_data = model(x_data) + np.random.normal(0, 1.5, len(x_data))

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='red', label='Model')
plt.scatter(x_data, y_data, color='black', label='Data')

equation = r'$y = 20.6 + 0.4945x$'
plt.text(10, 25, equation, fontsize=12, color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title(r'The graph for the equation $y = 20.6 + 0.4945x$ and the data can be plotted as shown below,')

plt.xlim(0, 60)
plt.ylim(0, 50)

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def model(x):
    return 49.358 * (x / (9.5075 + x))

x = np.linspace(0, 60, 1000)
y = model(x)

np.random.seed(42)
x_data = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])
y_data = model(x_data) + np.random.normal(0, 1.5, len(x_data))

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='red', label='Model')
plt.scatter(x_data, y_data, color='black', label='Data')

equation = r'$y = 49.358 \left( \frac{x}{9.5075 + x} \right)$'
plt.text(20, 25, equation, fontsize=12, color='black')

plt.xlabel('x')
plt.ylabel('y')
plt.title(r'The model $y = 49.358 \left( \frac{x}{9.5075 + x} \right)$ and the data can be plotted as shown below,')

plt.xlim(0, 60)
plt.ylim(0, 50)

plt.legend()

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def model(x):
    return -0.01606 * x**2 + 1.37788 * x + 11.7667

x = np.linspace(0, 60, 1000)
y = model(x)

np.random.seed(42)
x_data = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])
y_data = model(x_data) + np.random.normal(0, 1.5, len(x_data))

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='red', label='Model')
plt.scatter(x_data, y_data, color='black', label='Data')

equation = r'$y = -0.01606x^2 + 1.37788x + 11.7667$'
plt.text(5, 15, equation, fontsize=12, color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title(r'The graph for the equation $y = -0.01606x^2 + 1.37788x + 11.7667$ is shown below,')

plt.xlim(0, 60)
plt.ylim(0, 50)

plt.legend()
plt.grid(True)
plt.show()
```
